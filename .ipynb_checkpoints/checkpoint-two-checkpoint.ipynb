```python
corr = df.corr(numeric_only=True)
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corr, cmap="Greens",  annot=True)
# see if I can spot strong correlations or no correlation.
```

```python
# Looks like '% Silica Concentrate' is not correlated to any particular parameter.
# The same with '%Iron Feed'.
# I can spot some other correlations though.
# 'Ore pulp pH' has positive correlation with 'Starch Flow'
# 'Amina Flow' has strong positivev correlation with 'Ore Pulp Density'.
# 'Flotation Column 01 Air Flow' and 'Flotation Column 06 Air Flow' have negative correlation to 'Flotation Column 01 Level'
# 'Flotation Column 01 Air Flow' and 'Flotation Column 04 Air Flow' have stron positive correlation with 'Flotation Column 06 Air Flow'
# 'Flotation Column 04 Air Flow' has negative correlation with 'Flotation Column 05 Air Flow'
# 'Flotation Column 04 Air Flow' has positive correlation with ''Flotation Column 01 Air Flow''
```

## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
#The goal is to minimize Silica in final concentrate. I've sorted table on '% Silica Concentrate column' and plotted against other parameters.
fig, ax = plt.subplots(figsize=(20, 10))
df=df.sort_values('% Silica Concentrate')
plt.plot(df['% Silica Concentrate'],df['Ore Pulp pH'])
# plt.plot(df['Ore Pulp pH'])
# plt.plot(df['Ore Pulp Density'])
plt.xlim([0.75, 2])
plt.show()
```
